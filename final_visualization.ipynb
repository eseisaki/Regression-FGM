```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def plot_diagram(title, xlabel , ylabel,filename):

    plt.plot(plt1[:,1],plt1[:,0],'o-',linewidth=2)
    plt.plot(plt2[:,1],plt2[:,0],'o-',linewidth=2)


    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend(['FGM','GM'])
    plt.grid()
    plt.savefig("io_files/finalised/"+filename,dpi=150)
    plt.show()
```

```python
file="final_results.csv"
info=['algorithm', 'warmup', 'dataset', 'variable']
measurements=['sites', 'features', 'threshold', 'window','accuracy', 'rounds', 'traffic']

# Loading the experiments.
data = pd.DataFrame(pd.read_csv(file))
data

fixed_nw = data.query('dataset == "fixed" and warmup =="n"')[['sites', 'features', 'threshold', 'window','accuracy',
                                                           'rounds','traffic','algorithm', 'variable']]
drift_nw = data.query('dataset == "drift" and warmup =="n"')[['sites', 'features', 'threshold', 'window','accuracy',
                                                           'rounds','traffic','algorithm', 'variable']]
drift_w = data.query('dataset == "drift" and warmup =="y"')[['sites', 'features', 'threshold', 'window','accuracy',
                                                           'rounds','traffic','algorithm', 'variable']]
```
