```python
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def import_metric(file, name):
    df = pd.read_csv(file,names=[name,"time"])
    return df
```

```python
def plot_real_model():
    data = import_metric("io_files\inputs\drift_norms.csv","model norm")
    
    fig, ax = plt.subplots(figsize=(20, 5))
    
    ax.plot(data["time"], data["model norm"],linewidth=3) 
    ax.grid(True,linestyle='dashdot', color='grey')
    ax.tick_params(labelcolor='red', labelsize=15,pad=10,direction = 'inout', width=3)
    # set axis titles
    ax.set_xlabel("unit time",fontsize=20)
    ax.set_ylabel("model norm",fontsize=20)
    # set central titles
    ax.set_title( "Norm of real model (per pair)", size=25,pad =15)
```

```python
def plot_est_model(filename1, filename2):
    data = import_metric(filename1,"model norm")
    
    fig, ax = plt.subplots(figsize=(20, 5))
    
    ax.plot(data["time"], data["model norm"],linewidth=3) 
    ax.tick_params(labelcolor='red', labelsize=15,pad=10,direction = 'inout', width=3)
    
    rounds = import_metric(filename2,"rounds")
    for xc in rounds["time"]:
      ax.axvline(x=xc, color='gray', alpha=0.5)
        
    # set axis titles
    ax.set_xlabel("unit time",fontsize=20)
    ax.set_ylabel("model norm",fontsize=20)
    # set central titles
    ax.set_title( "Norm of estimate model (per pair)", size=25,pad =15)
    
    plt.ylim(1.5,2.5)
    
   
    
```

```python
def plot_regret(inName,outName):
    df1 = import_metric(inName,"real norm")
    df2 = import_metric(outName, "est norm")
    
    merged = pd.merge(df1,df2, on="time",how ="outer")
    merged['difference'] = abs(merged['real norm']-merged['est norm'])
    merged = merged[:-1]
    merged.fillna( method ='ffill', inplace = True) 
    
    fig, ax = plt.subplots(figsize=(20, 5))
    
    ax.plot(merged["time"], merged["difference"],linewidth=3) 
    ax.grid(True,linestyle='dashdot', color='grey')
    ax.tick_params(labelcolor='red', labelsize=15,pad=10,direction = 'inout', width=3)
    # set axis titles
    ax.set_xlabel("unit time",fontsize=20)
    ax.set_ylabel("regret",fontsize=20)
    # set central titles
    ax.set_title( "Regret - model error", size=25,pad =15)
    
    plt.ylim(0,2.5)
    
    
    
```

### Plot for Real Model Norm

```python
plot_real_model()
```

### Plot for Estimated Model Norm ( with depiction of rounds)

```python
filename1 = "io_files/gm/drift/nodes/output/k10_ft10_e5_win12000_ep3.csv"
filename2 = "io_files/gm/drift/nodes/rounds/k10_ft10_e5_win12000_ep3.csv"
plot_est_model(filename1,filename2)
```

### Plot for model error ( |real model norm - estimate model norm| )

```python
inName = "io_files/inputs/drift_norms.csv"
outName = "io_files/gm/drift/nodes/output/k10_ft10_e5_win12000_ep3.csv"
plot_regret(inName, outName)
```
